## SEL 2 — Variabel dan Tipe Data Dasar

Python mendukung beberapa tipe data dasar:
- **String (`str`)** — teks, diapit tanda kutip
- **Integer (`int`)** — bilangan bulat
- **Float (`float`)** — bilangan desimal
- **Boolean (`bool`)** — nilai `True` / `False`

`f-string` digunakan untuk menyisipkan variabel ke dalam teks dengan sintaks `f"...{variabel}..."`.

```python
# SEL 2: Variabel dan tipe data dasar
nama     = "Junior Dany Wibisono"     # String
nim      = "250401020098"              # String
angkatan = 2025                        # Integer
ipk      = 3.75                        # Float
aktif    = True                        # Boolean

print(f"Nama    : {nama}")
print(f"NIM     : {nim}")
print(f"Angkatan: {angkatan}")
print(f"IPK     : {ipk}")
print(f"Aktif   : {aktif}")
```

## SEL 3 — List dan Perulangan

**List** adalah kumpulan item yang terurut, ditulis dalam kurung siku `[ ]`.  
**`for` loop** digunakan untuk melakukan iterasi pada setiap elemen list.  
**`enumerate()`** mengembalikan indeks dan nilai sekaligus (parameter `1` berarti mulai hitung dari 1).

```python
# SEL 3: List dan perulangan
tools = ["Python", "Google Colab", "Jupyter Notebook", "GitHub"]

print("\nTools Data Science yang akan kita pelajari:")
for i, tool in enumerate(tools, 1):
    print(f"  {i}. {tool}")
```

## SEL 4 — Fungsi Sederhana

**Fungsi** (didefinisikan dengan `def`) adalah blok kode yang dapat digunakan kembali.  
Fungsi dapat menerima **parameter** dan mengembalikan nilai dengan `return`.  
Teks di dalam `"""..."""` disebut **docstring** — dokumentasi singkat tentang fungsi.

```python
# SEL 4: Fungsi sederhana
def perkenalan(nama, prodi, angkatan):
    """Fungsi untuk menampilkan perkenalan mahasiswa."""
    return f"Hai! Saya {nama} dari {prodi}, angkatan {angkatan}."

hasil = perkenalan("Junior Dany Wibisono", "Informatika", 2025)
print(hasil)
```

## Latihan Mandiri — Tools Tambahan & Motivasi
