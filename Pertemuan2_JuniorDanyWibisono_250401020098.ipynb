## 🔍 Q2 — Kolom dengan Missing Values

```python
print('=' * 55)
print('  Q2: ANALISIS MISSING VALUES PER KOLOM')
print('=' * 55)

# Hitung jumlah dan proporsi missing values
missing_count = df.isnull().sum()
missing_pct   = df.isnull().mean() * 100

# Buat DataFrame ringkasan
missing_df = pd.DataFrame({
    'Jumlah_Missing' : missing_count,
    'Proporsi_%'     : missing_pct.round(2)
}).sort_values('Jumlah_Missing', ascending=False)

# Kolom dengan missing values
ada_missing = missing_df[missing_df['Jumlah_Missing'] > 0].copy()
ada_missing['Status'] = ada_missing['Proporsi_%'].apply(
    lambda x: '🔴 KRITIS (>50%)' if x >= 50 else ('🟡 SEDANG (10-50%)' if x >= 10 else '🟢 MINOR (<10%)')
)

print('\n📊 Kolom DENGAN missing values:')
print(ada_missing.to_string())

# Kolom tanpa missing
tidak_missing = missing_df[missing_df['Jumlah_Missing'] == 0].index.tolist()
print(f'\n✅ Kolom TANPA missing values ({len(tidak_missing)} kolom):')
print(f'   {", ".join(tidak_missing)}')

# Total
total_missing = df.isnull().sum().sum()
total_cells   = df.shape[0] * df.shape[1]
print(f'\n📈 Total sel kosong  : {total_missing} dari {total_cells:,} sel ({total_missing/total_cells*100:.2f}%)')

print('\n📋 Rekomendasi Penanganan:')
print('   • Cabin    → Drop kolom (>77% kosong, tidak informatif)')
print('   • Age      → Imputasi dengan nilai median (±19.87% kosong)')
print('   • Embarked → Imputasi dengan modus / nilai terbanyak (0.22%)')
```

### 📝 Interpretasi Q2

Terdapat **3 kolom** yang memiliki missing values:

| Kolom | Missing | Proporsi | Rekomendasi |
|-------|---------|----------|-------------|
| `Cabin` | 687 | 77.10% | 🔴 Drop — terlalu banyak kosong |
| `Age` | 177 | 19.87% | 🟡 Imputasi dengan median |
| `Embarked` | 2 | 0.22% | 🟢 Imputasi dengan modus ('S') |

Kolom **`Cabin`** memiliki missing value paling parah (77.10%), sehingga biasanya didrop dalam analisis lanjutan. Penanganan missing values adalah langkah penting sebelum pemodelan Machine Learning.

## 🔍 Q3 — Persentase Penumpang yang Selamat

```python
print('=' * 55)
print('  Q3: PERSENTASE PENUMPANG YANG SELAMAT')
print('=' * 55)

# ─── Menggunakan NumPy ──────────────────────────────────
survived_arr  = df['Survived'].to_numpy()          # konversi ke NumPy array
n_total       = len(survived_arr)
n_selamat     = int(np.sum(survived_arr))          # sum nilai 1
n_meninggal   = n_total - n_selamat
pct_selamat   = np.mean(survived_arr) * 100        # mean = proporsi
pct_meninggal = 100 - pct_selamat

print(f'\n📊 Hasil Perhitungan (menggunakan NumPy):')
print(f'   Total penumpang       : {n_total:>5}')
print(f'   ✅ Selamat  (1)       : {n_selamat:>5}  →  {pct_selamat:.2f}%')
print(f'   ❌ Meninggal (0)      : {n_meninggal:>5}  →  {pct_meninggal:.2f}%')

# ─── Verifikasi dengan Pandas ───────────────────────────
print(f'\n📋 Verifikasi dengan Pandas (value_counts):')
vc     = df['Survived'].value_counts()
vc_pct = df['Survived'].value_counts(normalize=True) * 100
print(f'   Survived = 0 (Meninggal) : {vc[0]} penumpang  ({vc_pct[0]:.2f}%)')
print(f'   Survived = 1 (Selamat)   : {vc[1]} penumpang  ({vc_pct[1]:.2f}%)')

# ─── Statistik NumPy Array ──────────────────────────────
print(f'\n📐 Statistik Array Survived (NumPy):')
print(f'   np.sum  (total selamat) = {int(np.sum(survived_arr))}')
print(f'   np.mean (proporsi)      = {np.mean(survived_arr):.4f}')
print(f'   np.min                  = {int(np.min(survived_arr))}')
print(f'   np.max                  = {int(np.max(survived_arr))}')
```

### 📝 Interpretasi Q3

Dari **891 penumpang** dalam dataset:
- **342 orang (38.38%)** berhasil selamat.
- **549 orang (61.62%)** tidak selamat.

Lebih dari separuh penumpang (61.62%) tidak berhasil diselamatkan — menggambarkan betapa dahsyatnya tragedi tenggelamnya kapal Titanic pada April 1912. Angka ini menjadi baseline penting untuk mengevaluasi model prediksi Machine Learning nantinya.

## 🔍 Q4 — Penumpang Wanita Kelas 1: Jumlah & Rata-rata Usia

```python
print('=' * 55)
print('  Q4: ANALISIS PENUMPANG WANITA KELAS 1')
print('=' * 55)

# ─── Boolean Indexing ───────────────────────────────────
mask_wanita = df['Sex'] == 'female'
mask_kelas1 = df['Pclass'] == 1

wk1 = df[mask_wanita & mask_kelas1]   # filter wanita kelas 1

jumlah      = len(wk1)
selamat     = int(wk1['Survived'].sum())
pct_selamat = wk1['Survived'].mean() * 100
rata_usia   = wk1['Age'].mean()
min_usia    = wk1['Age'].min()
max_usia    = wk1['Age'].max()
med_usia    = wk1['Age'].median()

print(f"\n🎫 Filter Applied: Sex == 'female' AND Pclass == 1")
print()
print(f'   Jumlah penumpang wanita kelas 1 : {jumlah} orang')
print(f'   Jumlah yang selamat             : {selamat} orang')
print(f'   Tingkat keselamatan             : {pct_selamat:.2f}%')
print()
print(f'   Statistik Usia:')
print(f'   ├── Rata-rata (mean)  : {rata_usia:.2f} tahun')
print(f'   ├── Median            : {med_usia:.2f} tahun')
print(f'   ├── Termuda (min)     : {min_usia:.2f} tahun')
print(f'   └── Tertua (max)      : {max_usia:.2f} tahun')

print('\n📋 Contoh data wanita kelas 1 (5 baris pertama):')
display(wk1[['Name','Age','Survived','Fare','Embarked']].head())

# Perbandingan konteks
print('\n📊 Perbandingan Tingkat Keselamatan (Sex × Pclass) dalam %:')
tabel = df.groupby(['Pclass','Sex'])['Survived'].mean().mul(100).round(2).unstack()
print(tabel.to_string())
```

### 📝 Interpretasi Q4

Terdapat **94 penumpang wanita kelas 1** dalam dataset ini.
- **91 dari 94 orang (96.81%)** berhasil selamat — angka yang sangat tinggi.
- **Rata-rata usia** mereka adalah sekitar **34.61 tahun**.

Tingkat keselamatan hampir 97% ini mencerminkan dua faktor utama:
1. **Kebijakan evakuasi:** "Women and children first" diprioritaskan.
2. **Lokasi kabin:** Penumpang kelas 1 berada di dek atas, paling dekat dengan sekoci penyelamat.

## 🔍 Q5 — Tingkat Keselamatan per Kelas Penumpang (Groupby)

```python
print('=' * 55)
print('  Q5: TINGKAT KESELAMATAN PER KELAS PENUMPANG')
print('=' * 55)

# ─── Groupby dasar ──────────────────────────────────────
print('\n📊 [1] Groupby dasar: mean Survived per Pclass')
basic = df.groupby('Pclass')['Survived'].mean().mul(100).round(2)
print(basic.to_frame('% Selamat').to_string())

# ─── Groupby dengan agg() lengkap ───────────────────────
print('\n📊 [2] Groupby dengan .agg() — multi-fungsi:')
hasil = df.groupby('Pclass').agg(
    Total_Penumpang = ('PassengerId', 'count'),
    Jumlah_Selamat  = ('Survived'   , 'sum'),
    Pct_Selamat     = ('Survived'   , 'mean'),
    Rata_Usia       = ('Age'        , 'mean'),
    Rata_Fare       = ('Fare'       , 'mean')
).round(3)
hasil['Pct_Selamat']     = (hasil['Pct_Selamat'] * 100).round(2)
hasil['Jumlah_Selamat']  = hasil['Jumlah_Selamat'].astype(int)
print(hasil.to_string())

# ─── Kesimpulan ─────────────────────────────────────────
kelas_terbaik  = hasil['Pct_Selamat'].idxmax()
kelas_terburuk = hasil['Pct_Selamat'].idxmin()
print(f'\n🏆 Kelas TERTINGGI : Kelas {kelas_terbaik} → {hasil.loc[kelas_terbaik, "Pct_Selamat"]}% selamat')
print(f'📉 Kelas TERENDAH  : Kelas {kelas_terburuk} → {hasil.loc[kelas_terburuk, "Pct_Selamat"]}% selamat')

# ─── Pivot Table ────────────────────────────────────────
print('\n📊 [3] Pivot Table — % Survived (Pclass × Sex):')
pivot = pd.pivot_table(
    df,
    values  = 'Survived',
    index   = 'Pclass',
    columns = 'Sex',
    aggfunc = 'mean'
).mul(100).round(2)
print(pivot.to_string())
```

### 📝 Interpretasi Q5

Berdasarkan `groupby('Pclass')['Survived'].mean()`:

| Kelas | Total | Selamat | % Selamat |
|-------|-------|---------|----------|
| **1** | 216 | 136 | **62.96%** 🏆 |
| **2** | 184 | 87 | **47.28%** |
| **3** | 491 | 119 | **24.24%** 📉 |

**Kelas 1** memiliki tingkat keselamatan **tertinggi (62.96%)**, hampir 3× lebih tinggi dari kelas 3 (24.24%). Hal ini sangat berkorelasi dengan lokasi kabin (dek atas = akses sekoci lebih mudah) dan status sosial-ekonomi penumpang.

## 🔍 Q6 — Analisis Tambahan

```python
print('=' * 55)
print('  Q6A: TINGKAT KESELAMATAN BERDASARKAN KELOMPOK USIA')
print('=' * 55)

# Buat kolom kelompok usia
df_age = df.dropna(subset=['Age']).copy()
bins   = [0, 12, 17, 59, 100]
labels = ['Anak (<12)', 'Remaja (12-17)', 'Dewasa (18-59)', 'Senior (60+)']
df_age['KelompokUsia'] = pd.cut(df_age['Age'], bins=bins, labels=labels)

# Groupby kelompok usia
age_stats = df_age.groupby('KelompokUsia', observed=True).agg(
    Total    = ('Survived', 'count'),
    Selamat  = ('Survived', 'sum'),
    Pct      = ('Survived', 'mean')
)
age_stats['Pct']     = age_stats['Pct'].mul(100).round(2)
age_stats['Selamat'] = age_stats['Selamat'].astype(int)
age_stats.columns   = ['Total', 'Selamat', '% Selamat']
print()
print(age_stats.to_string())
```

```python
print('=' * 55)
print('  Q6B: TINGKAT KESELAMATAN BERDASARKAN UKURAN KELUARGA')
print('=' * 55)

# Buat kolom ukuran keluarga
df['UkKeluarga']  = df['SibSp'] + df['Parch']
bins2   = [-1, 0, 3, 6, 20]
labels2 = ['Sendiri (0)', 'Kecil (1-3)', 'Sedang (4-6)', 'Besar (7+)']
df['GrupKeluarga'] = pd.cut(df['UkKeluarga'], bins=bins2, labels=labels2)

# Groupby ukuran keluarga
fam_stats = df.groupby('GrupKeluarga', observed=True).agg(
    Total   = ('Survived', 'count'),
    Selamat = ('Survived', 'sum'),
    Pct     = ('Survived', 'mean')
)
fam_stats['Pct']     = fam_stats['Pct'].mul(100).round(2)
fam_stats['Selamat'] = fam_stats['Selamat'].astype(int)
fam_stats.columns   = ['Total', 'Selamat', '% Selamat']

print()
print(fam_stats.to_string())
print('\n📌 Penumpang dengan keluarga kecil (1-3 anggota) memiliki')
print('   peluang selamat lebih tinggi dibanding yang pergi sendiri.')
```

```python
print('=' * 55)
print('  Q6C: KORELASI TARIF TIKET (FARE) & KESELAMATAN')
print('=' * 55)

# Rata-rata Fare per status selamat
fare_stats = df.groupby('Survived')['Fare'].agg(['mean','median','min','max']).round(2)
fare_stats.index = ['❌ Meninggal (0)', '✅ Selamat (1)']

print()
print(fare_stats.to_string())

# Korelasi numerik
korelasi_fare = df[['Fare','Survived']].corr().iloc[0,1]
korelasi_age  = df[['Age','Survived']].dropna().corr().iloc[0,1]

print(f'\n📊 Korelasi Pearson:')
print(f'   Fare ↔ Survived : {korelasi_fare:.4f}  (positif lemah)')
print(f'   Age  ↔ Survived : {korelasi_age:.4f}  (negatif sangat lemah)')

print(f'\n💡 Kesimpulan:')
print(f'   Semakin tinggi tarif tiket, sedikit lebih besar kemungkinan selamat.')
print(f'   Fare sangat berkaitan dengan Pclass sebagai variabel utama.')
```

### 📝 Interpretasi Q6

**6A — Kelompok Usia:**
- **Anak-anak (<12 th)** memiliki tingkat keselamatan tertinggi (~59%) → prioritas "children first".
- **Dewasa (18–59 th)** memiliki tingkat terendah (~38%) karena merupakan kelompok terbesar dengan persaingan sekoci paling ketat.

**6B — Ukuran Keluarga:**
- **Keluarga kecil (1–3 anggota)** → tingkat selamat tertinggi (~58%) → saling membantu saat evakuasi.
- **Pergi sendiri** → hanya ~30% selamat → tidak ada yang membantu koordinasi.
- **Keluarga sangat besar (7+)** → sangat sulit dievakuasi bersama, tingkat selamat sangat rendah.

**6C — Tarif Tiket:**
- Rata-rata tarif penumpang yang selamat lebih tinggi → berkorelasi positif lemah dengan keselamatan (Fare berkaitan erat dengan `Pclass`).
